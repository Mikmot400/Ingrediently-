```python
from openfoodfacts import utils
import requests
def get_product(barcode, locale='Denmark'):
    """
    Return information of a given product.
    """
    url = utils.build_url(geography=locale,
                          service='api',
                          resource_type='product',
                          parameters=barcode)
    return utils.fetch(url)
```
